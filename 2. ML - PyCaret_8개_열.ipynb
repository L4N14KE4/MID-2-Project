```python
from pycaret.classification import *

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
diabates = pd.read_csv('diabates_full.csv', index_col=0) # index번호 없앰
diabates.head()
```

```python
# diabates['Age'] = diabates['Age'].astype(float)
# diabates['Glucose'] = diabates['Glucose'].astype(float)
# diabates['BMI'] = diabates['BMI'].astype(float)
# diabates['BloodPressure'] = diabates['BloodPressure'].astype(float)
# diabates['SkinThickness'] = diabates['SkinThickness'].astype(float)
# diabates['Insulin'] = diabates['Insulin'].astype(float)
# diabates['BMI'] = diabates['BMI'].astype(float)
# diabates['DiabetesPedigreeFunction'] = diabates['DiabetesPedigreeFunction'].astype(float)
```

```python
diabates = diabates.astype(float)
diabates['Outcome'] = diabates['Outcome'].astype(int)  # 'Outcome' int로 되돌리기
```

```python
diabates
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler

# 숫자형 변수들의 이름 리스트
numeric_columns = ['Pregnancies','Glucose',	'BloodPressure',	'SkinThickness',	'Insulin',	'BMI',	'DiabetesPedigreeFunction', 'Age']

# RobustScaler 사용 전처리기 정의
preprocessor = ColumnTransformer(
    transformers=[('numeric', RobustScaler(), numeric_columns)]
)
# 파이프라인에 추가
preprocessor_pipe = Pipeline(steps=[('preprocessor', preprocessor)]) # preprocessing-only

# fit & transform
preprocessor_pipe.fit(diabates)
diabates_transformed = preprocessor_pipe.transform(diabates)

# DataFrame로 변환 & 컬럼 이름 재지정
diabates_transformed = pd.DataFrame(diabates_transformed)
diabates_transformed.columns = ['Pregnancies','Glucose',	'BloodPressure',	'SkinThickness',	'Insulin',	'BMI',	'DiabetesPedigreeFunction', 'Age']
diabates_transformed['Outcome'] = diabates['Outcome']

diabates_transformed.head(3)
```

```python
# PyCaret setup 함수로 실험 환경 설정: 훈련 데이터와 목표 변수, 훈련 세트 크기 등 설정
model = setup(data=diabates_transformed,
              target='Outcome',
              train_size=0.7,
              # normalize=True,
              session_id=9)
```

```python
# Accuracy 기준 top 3 모델 선택
top_3_models = compare_models(sort='Accuracy',
                              n_select = 3) # Select top n models
```

```python
# 3개 모델 조합
blended = blend_models(estimator_list=top_3_models,
                       fold=10, # default
                       optimize='Accuracy',
                       method = 'hard')
```

```python
# 조합 모델 예측 수행
predict_model(blended)
```

```python
# 최종 모델 확정
final_blended = finalize_model(blended)
print(final_blended)
```

```python
# 최종 모델로 예측 수행
predict_model(final_blended)
```
